```python
#importing the time module
import time

#importing csv file containing 4000_words
import csv

#importing the random module
import random

#creating a variable 'words' to an empty list
words = []

def draw_hangman(tries):
    '''
    This function draws the hangman based on the number of attempts left.

    Parameters:
        tries (int): The number of attempts left.

    Returns:
        None
    '''
    if tries == 6:
        print('________ ')
        print('|  | ')
        print('| ')
        print('| ')
        print('| ')
        print('| ')
    elif tries == 5:
        print('________ ')
        print('|  | ')
        print('|  0 ')
        print('| ')
        print('| ')
        print('| ')
    elif tries == 4:
        print('________ ')
        print('|  | ')
        print('|  0 ')
        print('|  | ')
        print('| ')
        print('| ')
    elif tries == 3:
        print('________ ')
        print('|  | ')
        print('|  0 ')
        print('| /| ')
        print('| ')
        print('| ')
    elif tries == 2:
        print('________ ')
        print('|  | ')
        print('|  0 ')
        print('| /|\ ')
        print('| ')
        print('| ')
    elif tries == 1:
        print('________ ')
        print('|  | ')
        print('|  0 ')
        print('| /|\ ')
        print('| / ')
        print('| ')


def select_difficulty_mode():
    '''
    This function asks the user to select a difficulty mode for the game.

    The three options are: 'easy', 'medium', or 'hard'. The player's input is case-insensitive.

    Returns:
        str: The selected difficulty mode ('easy', 'medium', or 'hard').
    '''
    while True:
        mode = input('Please select the difficulty mode (easy, medium, hard): ').lower()
        if mode in ['easy', 'medium', 'hard']:
            return mode
        else:
            print('Invalid input. Please enter "easy", "medium", or "hard".')

def get_difficulty_mode(word_length):
    '''
    This Function determines the difficulty mode based on the word length.

    Parameters:
        word_length (int): The length of the word.

    Returns:
        str: The difficulty mode based on the word length ('easy', 'medium', or 'hard').
    '''
    if 2 <= word_length <= 4:
        return 'easy'
    elif 5 <= word_length <= 6:
        return 'medium'
    else:
        return 'hard'

#creating an empty list to store the words

#opening the CSV file and reading its contents
with open('4000_words.csv', newline='') as csvfile:
    #creating a CSV reader object
    reader = csv.reader(csvfile)
    #iterating over each row in the CSV file and appending the first column to the list
    words.extend(row[0] for row in reader)

#determining how many attempts the user gets
attempts = 7

while True:
    #welcoming the user
    name = input('What is your name?').capitalize()

    #using f-string to print the welcoming message
    print(f'Hello {name}. Welcome to Hangman!')

    #wait for 2 seconds
    time.sleep(2)
    print(f'You will have {attempts} attempts to guess the word. Good luck!')

    #wait for 3.5 seconds
    time.sleep(3.5)
    print('Before we begin...')

    #wait for 0.5 seconds
    time.sleep(1)

    #selecting the difficulty mode
    difficulty_mode = select_difficulty_mode()
    print(f'You selected {difficulty_mode} mode.')

    #wait for 1.5 seconds
    time.sleep(1.5)
    
    print('Let the game begin!')

    #wait for 1.5 seconds
    time.sleep(1.5)

    #filter words based on difficulty mode
    filtered_words = [word for word in words if get_difficulty_mode(len(word)) == difficulty_mode]

    #selecting a random word from the filtered words list
    word = random.choice(filtered_words)

    #creating a variable with an empty value
    guesses = ''

    #creating a variable with the guessed letters
    guessed = [guesses]

    #determining the maximum number of tries the user has
    tries = 7

    #CREATING A WHILE LOOP
    #checking if the number of tries is more than 0
    while tries > 0:    
        # print hangman
        draw_hangman(tries)

        #making a counter that starts with 0
        counter = 0             

        #for every letter in the randomly selected word   
        for letter in word:      
            #see if the letter is in the player's guess
            if letter in guesses:    
                #print the letter, on the same line
                print(letter,end='')
            else:
                #if letter is not found in the word found, print a dash
                print('_ ',end=''),     
                #and increase the counter by one
                counter += 1    

        #if counter is equal to zero
        if counter == 0:
            print('. Congratulations! You have guessed the word.')
            #exit the script
            break            

        #ask the user to guess a letter
        guess = input('Guess a letter:') 

        #ensures the user guesses single letters only
        if len(guess) != 1 or not guess.isalpha():
            print('Invalid input. Please enter a single letter.')
            continue
    
        #set the players guess to guesses
        guesses += guess                    

        #if the guess is not found in the randomly selected word
        if guess not in word:  
            #if the guess has already been entered
            if guess in guessed:
                print('You have already guessed this letter!')
                continue
 
            #tries counter decreases by 1. same as 'tries = tries - 1'
            tries -= 1        
 
            #print wrong
            print('Wrong!')  
 
            #number of guesses left
            print('You have', tries, 'more guesses' )

        else:
            if guess in word:
                if guess in guessed:
                    print('You have already guessed this letter!')
                    continue

        #populates the player's guess into the list of guessed letters
        guessed.append(guess)
    
        #if the turns are equal to zero
        if tries == 0:           
            #print 'You Lose'
            print('________ ')
            print('|  | ')
            print('|  0 ')
            print('| /|\ ')
            print('| / \ ')
            print('| ')
            print('You Lose. Better luck next time!')
            break

    #ask the user if they would like to play again
    play_again = input('Do you want to play again? (yes/no): ')

    #for as long as the player's input doesn't match 'yes' or 'no'
    while play_again.lower() not in ['yes', 'no']:
        print('Invalid input. Please enter "yes" or "no".')
        play_again = input('Do you want to play again? (yes/no): ').lower()

    #if no, end game. since its a loop, 'yes' will start the game again.
    if play_again.lower() == 'no':
        print('Thank you for playing!')
        break
```
